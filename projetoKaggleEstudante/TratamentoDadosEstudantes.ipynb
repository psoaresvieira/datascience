```python
import pandas as pd
import seaborn as srn
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('c:\\Users\mvspb\student-scores.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.rename(columns={'id':'Id', 'first_name':'Primeiro Nome', 'last_name': 'Último Nome','email': 'Email', 'gender': 'Gênero','part_time_job': 'Trabalha meio Período', 'absence_days': 'Total de Dias de Ausência','extracurricular_activities': 'Atividades Extracurriculares','weekly_self_study_hours': 'Total de Horas Semanais Estudadas','career_aspiration': 'Aspiração de Carreira','math_score': 'Nota em Matemática','history_score': 'Nota em História','physics_score': 'Nota em Física','chemistry_score': 'Nota em Química','biology_score': 'Nota em Biologia','english_score': 'Nota em Inglês','geography_score': 'Nota em Geografia'}, inplace=True)
df.head()
df.to_csv('novo_arquivo.csv', index=False)
```

```python
df.head()
```

```python
agrupado = df.groupby('Genero').size()

agrupado.plot.bar(color = 'blue')
plt.ylabel("Total Aluno")
plt.title("Total de Aluno por genero")

agrupado
```

```python
agrupado_atv_extra = df.groupby('Atividades Extracurriculares').size()

agrupado_atv_extra.plot.bar(color = "green")
plt.ylabel('Total Alunos')
plt.title('Total de Alunos que fizeram ou não Atividades Extras')
```

```python
agrupado1 = df.groupby('Genero')[['Total Horas Semanais Estudadas']].mean()
agrupado1.plot(kind = 'bar')
plt.title('Total de Horas Semanais Estudas por Gênero')
plt.legend(ncol=1, loc='upper right', bbox_to_anchor=(1.6, 1))
```
